```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from scipy.interpolate import interp1d
from math import floor
from scipy.ndimage.filters import gaussian_filter1d

%matplotlib inline
```

```python
#Sensitivity Sony DXC 930 https://nae-lab.org/~rei/research/cs/zhao/database.html 
wale=[]
R=[]
G=[]
B=[]
with open('cam_s.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            wale.append(float(row[0]))
            R.append(float(row[1]))
            G.append(float(row[2]))
            B.append(float(row[3]))
            line_count += 1
    print(f'Processed {line_count} lines.')
    
plt.plot(wale,R,'r',label='red')
plt.plot(wale,G,'g', label='green')
plt.plot(wale,B,'b',label='blue')
plt.title(" Sony DXC 930")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Spectral Sensitivity S($\lambda$)")
plt.legend()
plt.savefig('cam_s.png')
```

```python
#Interpolation functions
nsenR = interp1d(wale, R, kind='cubic')
nsenG = interp1d(wale, G, kind='cubic')
nsenB = interp1d(wale, B, kind='cubic')

nwh = interp1d(mwh, imwh, kind='cubic')

senR = np.array(nsenR(mwh))
senG = np.array(nsenG(mwh))
senB = np.array(nsenB(mwh))
```

```python
fig = plt.figure()
print("R","G","B")
for i in range(1,25):
    nx = eval('m' + str(i))
    ny = eval('im' + str(i))
    nids = interp1d(nx,ny,kind='cubic')
    ids = np.array(nids(mwh))
    DR = floor(255*sum(np.multiply(np.multiply(imwh,ids),senR))/sum(np.multiply(imwh,senR)))
    DG = floor(255*sum(np.multiply(np.multiply(imwh,ids),senG))/sum(np.multiply(imwh,senG)))
    DB = floor(255*sum(np.multiply(np.multiply(imwh,ids),senB))/sum(np.multiply(imwh,senB)))
    
    if i>=1 and i<7:
        x=1.85
        if i==1:
            x=4
        elif i>=3 or i<6:
            2.5
    elif i>=7 and i<11:
        x=2
        if i==9:
            x=3
        if i==10:
            x=4.3
    elif i>=11 and i<13:
        x=1.68
    elif i>=13 and i<19:
        x=2.5
        if i==15:
            x=4.5
        if i==16:
            x=1.3
    else:
        x=1
    
    print(DR,DG,DB)
    DR = DR*x
    DG = DG*x
    DB = DB*x
    
    fin = np.array([[[  DR,  DG,    DB],
        [DR,  DG,    DB]],

       [[  DR,  DG,    DB],
        [DR,  DG,    DB]]], dtype='uint8')
    
    
 
    globals()['sub%s' % i] = fig.add_subplot(4,6,i) # instead of plt.subplot(2, 2, 1)
    globals()['sub%s' % i].imshow(fin)
    globals()['sub%s' % i].axis('off')


plt.savefig('Macbeth_Recreate.png')    
```
